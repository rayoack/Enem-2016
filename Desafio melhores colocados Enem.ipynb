```python
# IMPORTANDO PANDAS
import pandas as pd
```

```python
# CRIANDO DATAFRAME
df = pd.read_csv('train.csv', sep=',')
```

# Calculando a média das notas

Na internet cada site fala um modo diferente para calcular a média das notas do enem com os pesos, mas acredito que o modod padrão seja multiplicar cada nota pelo seu peso, somar todas elas, e dividir pela soma dos pesos, então vou seguir esse padrão.

Vamos separar as colunas das notas e do numero da inscrição em um outro objeto.

Depois calcular a média e inserir em uma nova coluna, e lgo após salvar os melhores colocados e um arquivo CSV separado.

#### Os pesos determinado pelo site são:

- matemática: 3
- ciências da natureza: 2
- linguagens e códigos: 1.5
- ciências humanas: 1
- redação: 3

```python
# Separando colunas em um outro objeto
df_notas = df[['NU_INSCRICAO',
               'NU_NOTA_MT',
               'NU_NOTA_CN',
               'NU_NOTA_LC',
               'NU_NOTA_CH',
               'NU_NOTA_REDACAO']]
```

```python
df_notas.head()
```

```python
# Calculando a média das notas e inserindo em uma nova coluna chamada 'NOTA_FINAL'
df_notas['NOTA_FINAL'] = (
    df_notas['NU_NOTA_MT'] * 3 +
    df_notas['NU_NOTA_CN'] * 2 +
    df_notas['NU_NOTA_LC'] * 1.5 +
    df_notas['NU_NOTA_CH'] * 1 +
    df_notas['NU_NOTA_REDACAO'] * 3) / (3 + 2 + 1.5 + 1 + 3)
```

```python
df_notas.head()
```

```python
# Os 20 melhores colocados
df_notas[['NU_INSCRICAO', 'NOTA_FINAL']].sort_values(by = ['NOTA_FINAL'], ascending = False).head(20)
```

```python
# Salvando os melhores 20 colocados em um arquivo CSV separado
df_notas[['NU_INSCRICAO', 'NOTA_FINAL']].sort_values(
    by = ['NOTA_FINAL'], ascending = False
).head(20).to_csv("answer.csv",index=False)
```
